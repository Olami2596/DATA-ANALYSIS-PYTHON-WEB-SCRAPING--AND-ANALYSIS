## Car Sales on AA

```python
#importing all necessary libraries 
import pandas as pd
import numpy as np
import requests
import json
from bs4 import BeautifulSoup
from csv import writer
import matplotlib.pyplot as plt 
import seaborn as sns
```

#### Data gathering

```python
#creating a for loop to extract the needed information from the website
#the vehicle id was needed so as to be able to get the complete information for each car which is located in another page on the website
#the for i in range(1, 61) was used due to the number of the cars needed for the analysis
vehicle_id = []
cars = []
for i in range(1, 61):
    url = f"https://www.theaa.com/used-cars/displaycars?fullpostcode=B11AD&travel=2000?AC={i}"
    page = requests.get(url, headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"})
    soup = BeautifulSoup(page.content, 'html.parser')
    lists = soup.find_all('div', class_ = "vl-item clearfix")
    
    for list in lists:
        title = list.find('span', class_ = "make-model-text").text
        car_id = list.find('button', class_ = "shortlist-btn" )
        id = car_id["data-table-uid"]
        price_1 =list.find('strong', class_ = "total-price strong-inline new-transport--bold").text.replace('\n', '')
        price = price_1[1:7]
        vehicle_id.append(id)
        cars.append([title,id,price])
```

```python
# a for loop was created to get the location,mileage,year,fuel_type,transmission,body_type,color,doors,engine_size,co2_transmission from the website and it was all stored in the list over
# the for k in vehicle_id was used because the vehicle_id contains the id for each car so as to enable acess to the complete information for each car
over  = [] 
for k in vehicle_id:
    url_1 = f"https://www.theaa.com/used-cars/cardetails/39-{k}"
    page_1 = requests.get(url_1, headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"})
    soup_1 = BeautifulSoup(page_1.content, 'html.parser')
    records = soup_1.find('div', class_="site-wrap")

    location = records.find('h3').text
    mileage = records.find_all('span', class_ = 'vd-spec-value')[0].text
    year = records.find_all('span', class_ = 'vd-spec-value')[1].text
    fuel_type = records.find_all('span', class_ = 'vd-spec-value')[2].text
    transmission = records.find_all('span', class_ = 'vd-spec-value')[3].text
    body_type = records.find_all('span', class_ = 'vd-spec-value')[4].text
    color = records.find_all('span', class_ = 'vd-spec-value')[5].text
    doors = records.find_all('span', class_ = 'vd-spec-value')[6].text
    engine_size = records.find_all('span', class_ = 'vd-spec-value')[7].text
    co2_transmission = records.find_all('span', class_ = 'vd-spec-value')[8].text
    over.append([location,mileage,year,fuel_type,transmission,body_type,color,doors,engine_size,co2_transmission])
```

```python
over
```

```python
# a for loop was created to get the review count and the rating for each car
review  = []
for i in vehicle_id:
    url_1 = f"https://www.theaa.com/used-cars/cardetails/39-{i}"
    page_1 = requests.get(url_1, headers = {'User-Agent': "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/74.0.3729.169 Safari/537.36"})
    soup_1 = BeautifulSoup(page_1.content, 'html.parser')
    records = soup_1.find('div', class_="site-wrap")
    rating_review = records.find('div', class_ = 'score')
    if rating_review:
        review_count = rating_review.find_all('div')[0]['content']
        rating = rating_review.find_all('div')[2]['content']
        review.append([review_count,rating])
    else:
        # append default values if rating_review is not found
        review.append(['0','0'])

```

```python
review
```

```python
#the three lists cars, over and review were all converted into dataframes
df_1 = pd.DataFrame(cars, columns=["title","id","price"])
df_2 = pd.DataFrame(over, columns=['location','mileage','year','fuel_type', 'transmission', 'body type', 'color', 'doors', 'engine_size', 'co2_transmission'])
df_3 = pd.DataFrame(review, columns=['review_count', 'rating'])
```

```python
#the dataframes were joined together 
df = pd.concat([df_1, df_2,df_3], axis="columns")
df
```

```python
#converting the dataframe into csv format
df.to_csv('cars.csv')
```

```python
#making a copy of the dataset before any wrangling procedure is carried out
df_copy = df.copy()
df_copy
```

```python
#creating a function to remove the comma's in the gdp column and convert the column into int 
def comma_remover(x):
    return int(x.replace(',', ''))
```

```python
df_copy['price'] = df_copy['price'].apply(comma_remover)
df_copy['mileage'] = df_copy['mileage'].apply(comma_remover)
```

```python
df_copy
```

```python
#removing the L at the back of the values in the engine size column and the g/km at the back of the values in the co2_transmission column
df_copy['engine_size'] = df_copy['engine_size'].str.strip(' L')
df_copy['co2_transmission'] = df_copy['co2_transmission'].str.strip(' g/km')
```

```python
df_copy
```

```python
#converting columns from object into their appropriate format
df_copy[['year', 'doors', 'co2_transmission','review_count']] = df_copy[['year', 'doors', 'co2_transmission','review_count']].astype(int)
df_copy[['engine_size', 'rating']] = df_copy[['engine_size', 'rating']].astype(float)
```

```python
df_copy.info()
```

```python
#dropping all the null values
df_copy = df_copy.dropna()
```

```python
#dropping the id column because it is not needed for the analysis
df_copy = df_copy.drop('id', axis = 1)
```

#### Questions

1. Calculate the total car sales based on the year feature 

```python
df_total_sales_by_year = df_copy.groupby('year')['price'].sum().reset_index()
df_total_sales_by_year
```

2. Compare car sales on their transmission features 

```python
df_total_sales_by_transmission = df_copy.groupby('transmission')['price'].sum().reset_index()
df_total_sales_by_transmission
```

Cars with manual transmission generated more sales than cars with automatic transmission

3. What are the most popular car sales based on the Body Type?

```python
df_total_sales_by_body_type = df_copy.groupby('body type')['price'].sum().reset_index()
df_total_sales_by_body_type
```

```python
df_rank_body_type = df_total_sales_by_body_type.sort_values('price', ascending=False).reset_index(drop=True)
df_rank_body_type
```

The most popular car sales based on body type is the Hatchback followed by the van then the saloon and the least popular is the estate
